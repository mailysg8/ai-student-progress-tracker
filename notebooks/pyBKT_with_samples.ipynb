```python
import pandas as pd
from pyBKT.models import Model
import sys
sys.path.append('..')
from notebooks.src.split import split
```

## Data

```python
initial_data = pd.read_excel('../data/raw/Stellar_edu_MDS_ap_stats_dataset - v1.9.xlsx', sheet_name='Student_Observations')
sample_50 = pd.read_csv('../data/sample/sample_50.csv')
sample_100 = pd.read_csv('../data/sample/sample_100.csv')
```

```python
grouped_data = pd.read_csv('../data/processed/full_grouped_data.csv')
grouped_sample_50 = pd.read_csv('../data/sample/grouped_sample_50.csv')
grouped_sample_100 = pd.read_csv('../data/sample/grouped_sample_100.csv')
```

## Building models

### Original data

#### Initial data

```python
initial_train, initial_test, initial_val = split(initial_data)
```

```python
initial_model = Model(seed = 42, num_fits = 1)
initial_model.fit(data = initial_train)
```

#### 50 student sample

```python
fifty_train, fifty_test, fifty_val = split(sample_50)
```

```python
fifty_model = Model(seed = 42, num_fits = 1)
fifty_model.fit(data = fifty_train)
```

#### 100 student sample

```python
hundred_train, hundred_test, hundred_val = split(sample_100)
```

```python
hundred_model = Model(seed = 42, num_fits = 1)
hundred_model.fit(data = hundred_train)
```

### Grouped data

#### Initial data

```python
group_initial_train, group_initial_test, group_initial_val = split(grouped_data, 'modeling_kc_id')
```

```python
group_initial_model = Model(seed = 42, num_fits = 1)
group_initial_model.fit(data = group_initial_train)
```

#### 50 student sample

```python
group_fifty_train, group_fifty_test, group_fifty_val = split(grouped_sample_50, 'modeling_kc_id')
```

```python
group_fifty_model = Model(seed = 42, num_fits = 1)
group_fifty_model.fit(data = group_fifty_train)
```

#### 100 student sample

```python
group_hundred_train, group_hundred_test, group_hundred_val = split(grouped_sample_100, 'modeling_kc_id')
```

```python
group_hundred_model = Model(seed = 42, num_fits = 1)
group_hundred_model.fit(data = group_hundred_train)
```

## Getting parameters

```python
initial_params = initial_model.params().reset_index()
fifty_params = fifty_model.params().reset_index()
hundred_params = hundred_model.params().reset_index()
```

```python
group_initial_params = group_initial_model.params().reset_index()
group_fifty_params = group_fifty_model.params().reset_index()
group_hundred_params = group_hundred_model.params().reset_index()
```

## Getting predictions

```python
initial_preds = initial_model.predict(data=initial_test)
```

```python
fifty_preds = fifty_model.predict(data=fifty_test)
```

```python
group_fifty_preds = group_fifty_model.predict(data=group_fifty_test)
```

```python
hundred_preds = hundred_model.predict(data=hundred_test)
```

## Evaluation

```python
initial_rmse = initial_model.evaluate(data = initial_val)
initial_auc = initial_model.evaluate(data = initial_val, metric = 'auc')
print("Training RMSE: %f" % initial_rmse)
print("Training AUC: %f" % initial_auc)
```

```python
group_initial_rmse = group_initial_model.evaluate(data = group_initial_val)
group_initial_auc = group_initial_model.evaluate(data = group_initial_val, metric = 'auc')
print("Training RMSE: %f" % group_initial_rmse)
print("Training AUC: %f" % group_initial_auc)
```

```python
fifty_rmse = fifty_model.evaluate(data = fifty_val)
fifty_auc = fifty_model.evaluate(data = fifty_val, metric = 'auc')
print("Training RMSE: %f" % fifty_rmse)
print("Training AUC: %f" % fifty_auc)
```

```python
group_fifty_rmse = group_fifty_model.evaluate(data = group_fifty_val)
group_fifty_auc = group_fifty_model.evaluate(data = group_fifty_val, metric = 'auc')
print("Training RMSE: %f" % group_fifty_rmse)
print("Training AUC: %f" % group_fifty_auc)
```

```python
hundred_rmse = hundred_model.evaluate(data = hundred_val)
hundred_auc = hundred_model.evaluate(data = hundred_val, metric = 'auc')
print("Training RMSE: %f" % hundred_rmse)
print("Training AUC: %f" % hundred_auc)
```

```python
group_hundred_rmse = group_hundred_model.evaluate(data = group_hundred_val)
group_hundred_auc = group_hundred_model.evaluate(data = group_hundred_val, metric = 'auc')
print("Training RMSE: %f" % group_hundred_rmse)
print("Training AUC: %f" % group_hundred_auc)
```

## Summary

```python
print("Evaluation summary: ")
print("")
print("")
# RMSE
results = {
    25:  {"normal": initial_rmse, "grouped": group_initial_rmse},  
    50:  {"normal": fifty_rmse, "grouped": group_fifty_rmse},
    100: {"normal": hundred_rmse, "grouped": group_hundred_rmse},
}

header = f"{'Student Size':>12} | {'RMSE Normal':>12} | {'RMSE Grouped':>13}"
print("RMSE — Normal vs Grouped Data")
print("-" * len(header))
print(header)
print("-" * len(header))
for size, vals in results.items():
    n = f"{vals['normal']:.4f}"  if vals["normal"]  is not None else "—"
    g = f"{vals['grouped']:.4f}" if vals["grouped"] is not None else "—"
    print(f"{size:>12} | {n:>12} | {g:>13}")
print("-" * len(header))

results = {
    25:  {"normal": initial_auc, "grouped": group_initial_auc},  
    50:  {"normal": fifty_auc, "grouped": group_fifty_auc},
    100: {"normal": hundred_auc, "grouped": group_hundred_auc},
}
print("")
print("")
# AUC
header = f"{'Student Size':>12} | {'AUC Normal':>12} | {'AUC Grouped':>13}"
print("AUC — Normal vs Grouped Data")
print("-" * len(header))
print(header)
print("-" * len(header))
for size, vals in results.items():
    n = f"{vals['normal']:.4f}"  if vals["normal"]  is not None else "—"
    g = f"{vals['grouped']:.4f}" if vals["grouped"] is not None else "—"
    print(f"{size:>12} | {n:>12} | {g:>13}")
print("-" * len(header))
```
